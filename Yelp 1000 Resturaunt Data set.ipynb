# Fetching the Businesses

```python
while offset <= 1000:
    group = json.loads(requests.get((search_url + '?' +
             'latitude=' + Latitude + '&' +
             'longitude=' + Longitude + '&' +
             'term=' + term + '&' +
             'locale=' + locale + '&' +
             'offset=' + str(offset)), headers=headers).content.decode())
    for record in group['businesses']:
        dataset['businesses'].append(record)
    offset += len(group['businesses'])
    if offset % 100 == 0:
        print(str(int(offset/10)), 'percent complete')
    if offset/10 == 100:
        break
```

### Adding the data

```python
Resturaunts = {}
```

```python
for business in dataset['businesses']:
    print(business['name'])
    Resturaunts[business['name']] = json.loads(requests.get((business_uri + business['id'] + '?' + 'locale=' + locale), headers=headers).content.decode())
    Resturaunts[business['name']]['reviews'] = json.loads(requests.get((business_uri + business['id'] + '/reviews' + '?' + 'locale=' + locale), headers=headers).content.decode())
```

Storing Unique Business names in Dictionary format does not allow for duplicate entries so we have around 900 businesses rather than 1000 but it will work for our purposes.
